```python
# cau 1
import pandas as pd

data = pd.read_csv('P4AI_BT1.csv')
data
```

```python
# cau 2
df1 = data.copy()
df1['sepal.length'].fillna(round(data['sepal.length'].mean()), inplace=True)
df1['sepal.width'].fillna(round(data['sepal.width'].mean()), inplace=True)
df1['petal.length'].fillna(round(data['petal.length'].mean()), inplace=True)
df1['petal.width'].fillna(round(data['petal.width'].mean()), inplace=True)
df1['variety'].fillna(df1['variety'].mode()[0], inplace=True)
df1
```

```python
# cau 4
df2 = df1.copy()
slength_min = df2['sepal.length'].min()
slength_max = df2['sepal.length'].max()

swidth_min = df2['sepal.width'].min()
swidth_max = df2['sepal.width'].max()

plength_min = df2['petal.length'].min()
plength_max = df2['petal.length'].max()

pwidth_min = df2['petal.width'].min()
pwidth_max = df2['petal.width'].max()

df2['sepal.length'] = (df2['sepal.length'] - slength_min)/(slength_max - slength_min)
df2['sepal.width'] = (df2['sepal.width'] - swidth_min)/(swidth_max - swidth_min)
df2['petal.length'] = (df2['petal.length'] - plength_min)/(plength_max - plength_min)
df2['petal.width'] = (df2['petal.width'] - pwidth_min)/(pwidth_max - pwidth_min)

df2
```

```python
# cau 5
df3 = df2.copy()
df3['variety'].unique()
dumies = pd.get_dummies(df2['variety'])
print(dumies)
df3 = pd.concat([df3, dumies], axis=1)
df3.drop('variety', axis=1, inplace=True)
df3
```

```python
# cau 6
df4 = df2.copy()
df4['variety'].replace({"Setosa": 0, "Versicolor": 1, "Virginica": 2}, inplace=True)
Train = df4.drop("variety", axis=1)
Test = df4["variety"]
display(Train)
display(Test)
tp = int(0.9 * len(Train))
x_train = Train[:tp]
y_train = Test[:tp]
x_test = Train[tp:]
y_test = Test[tp:]

from sklearn.linear_model import RidgeClassifier

reg = RidgeClassifier().fit(x_train, y_train)
pred = reg.predict(x_test)
print(reg.score(x_test, y_test))
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

data = pd.read_csv('P4AI_BT1.csv')
df1 = data.copy()
df1['sepal.length'].fillna(round(data['sepal.length'].mean()), inplace=True)
df1['sepal.width'].fillna(round(data['sepal.width'].mean()), inplace=True)
df1['petal.length'].fillna(round(data['petal.length'].mean()), inplace=True)
df1['petal.width'].fillna(round(data['petal.width'].mean()), inplace=True)
df1['variety'].fillna(df1['variety'].mode()[0], inplace=True)
df1['variety'].replace({"Setosa": 0, "Versicolor": 1, "Virginica": 2}, inplace=True)
# Chọn các cột dữ liệu để phân cụm
# X = df1.iloc[:, 0:5]
X = df1.copy()
# Xác định số cụm k và áp dụng K-Means clustering
kmeans = KMeans(n_clusters=3)
kmeans.fit(X)

# Trực quan hóa kết quả phân cụm bằng biểu đồ scatter plot
fig, ax = plt.subplots(figsize=(8, 6))
ax.scatter(X.iloc[:, 0], X.iloc[:, 2], c=kmeans.labels_)
ax.set_xlabel('Column 1')
ax.set_ylabel('Column 2')
ax.set_title('K-Means Clustering')
plt.show()
```
